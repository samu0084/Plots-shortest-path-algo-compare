```python
import csv
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import collections
from collections import defaultdict
import math
```

```python
def read_csv(path):
    rows = []
    with open(path) as file:
        reader = csv.reader(file)
        for row in reader:
            rows.append(row)
    return rows
```

```python
def csv_to_list(csv_rows, row_index):
    lst = []
    for row in csv_rows:
        lst.append(int(row[row_index]))
    return lst
```

```python
def index_bound_row(x):
    return {
        50: 0,
        100: 1,
        150: 2,
        250: 3,
        400: 4, 
        800: 5,
        1000: 6,
        1800: 7,
        2800: 8,
        4600: 9,
        7400: 10,
        12000: 11,
        19400: 12,
        31400: 13,
        50800: 14
    }[x]
```
